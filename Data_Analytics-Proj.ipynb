```python
import numpy as np
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('BigBasket Products.csv')
```

```python
# Removing Index Column 

df.drop('index',axis=1,inplace=True)
```

```python
# Top and Least sold Products. (Frequency)

df['product'].value_counts()
```

```python
# Mesauring Discount on a Certain item.

df['discount'] = df.market_price - df.sale_price
```

```python
df
```

```python
# Find Missing or Null Values.

df.isnull().sum()
```

```python
df['product']  = df['product'].fillna('Unknown')
```

```python
df.brand = df.brand.fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
# Finding the index of missing values under sales price.  (Data Filteration).

df[df.sale_price.isnull()]
```

```python
# Filling Sale Price null values with Market Price Values.

df.sale_price = df.sale_price.fillna(df.market_price)
```

```python
df.isnull().sum()
```

```python
df[1718:1725]
```

```python
df.rating = df.rating.fillna(df.rating.median())
```

```python
df.description = df.description.fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.discount = df.discount.fillna(0)
```

```python
# Remove Unwanted Data. (Mistaked Data)

df.sale_price = np.where(df.sale_price>15000,df.market_price,df.sale_price)
```
